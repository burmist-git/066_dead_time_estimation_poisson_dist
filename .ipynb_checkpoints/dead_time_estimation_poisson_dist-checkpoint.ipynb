```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
number_of_events=200000
```

```python
time_arr = np.random.uniform(0, 1e+9, int(number_of_events))
```

```python
time_arr=np.sort(time_arr)
```

```python
time_arr_roll=np.roll(time_arr,shift=1)
```

```python
dt=time_arr-time_arr_roll
```

```python
dt=dt[1:]
```

```python
plt.hist(dt, bins=100, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('dt, ns')
plt.ylabel('Frequency')
plt.title('delta time between two consecutive events')

plt.yscale('log')

# Show the plot
plt.show()
```
